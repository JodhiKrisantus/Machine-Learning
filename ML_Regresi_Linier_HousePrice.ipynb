# REGRESI Linier
Untuk Melihat hubungan antara 2 variable. Untuk Membuat Estimasi, dan melakukan prediksi
- Contoh : Hubungan Antara Lama belajar, terhadap Nilai Kuliah
- Variable X = Variable Independen (yg mempengaruhi)
- Variable y = Variable Dependent (yg Dipengaruhi)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

Terdapat Data luas tanah dan Harga
- Kita coba prediksi apakah Harga dipengaruhi oleh Luas Tanah

```python
df = pd.read_csv('https://raw.githubusercontent.com/rasyidev/well-known-datasets/main/house_prices.csv')
df.head()
```

```python
df.shape # Terdapat 1326 baris, dan 2 kolom
```

## Mengambil variable X dan y dan memasukkan kedalam list
Lakukan Slicing untuk : 
- X(yg mempengaruhi) = lot area(Luas Tanah)
- y (yg dipengaruhi) = Price (Harga)

```python
from re import X
X = df.LotArea[:,np.newaxis] #Lot Area, Kita jadikan 2 dimensi (dibutuhkan 2 dimensi untuk regressor.fit, dan regressor.Predict)
y = df.SalePrice#Price
print(X)
print(y)
```

```python
# plt.scatter(data=df, x=df.LotArea, y=df.SalePrice)
plt.scatter(x=df.LotArea,y=df.SalePrice)
plt.xlabel("LotArea")
plt.ylabel("Price")
plt.title("Grafik Luas Tanah Terhadap Harga")
```

## Memecah Data Training Dan Data Testing
- Membagi data :
- 80 % untuk Training
- 20 % untuk Testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 0 )
X_train
```

## Fitting (TRAINING) Metode Regresi Linier
- Melakukan Fitting Simple Linier Regression pada Train set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

##  Training Process
Memprediksi X_train / Lot Area (Luas Tanah)
- machine learning akan membuat suatu model (y_Prediksi), yang ditunjukkan oleh Garis Merah pada saat divisualisasi

```python
y_prediksi =  regressor.predict(X_test)
y_prediksi
```

# Visualisasi X train, Terhadap Y_prediksi

```python
plt.figure(figsize=(12,8))
plt.scatter(X_train,y_train,color='b')
plt.plot(X_test,y_prediksi,color='r')
plt.title("Prediksi Harga, Berdasarkan Lot Area")
plt.xlabel("Luas Tanah")
plt.ylabel("Harga")
```

```python
# BILA INGIN MELIHAT DATA PREDIKSI SAJA 
plt.figure(figsize=(12,8))
plt.scatter(X_test,y_prediksi)
```

# Visualisasi Data Test
Kita memakai 20% data yang belum di training sebagai data testing yaitu : **X_test, dan y_test**
- Untuk Melihat, apakah dari 20% data testing(Data Biru) sudah sesuai / mirip dengan data prediksi (Garis Merah)

```python
plt.figure(figsize=(12,8))
plt.scatter(X_test,y_test,color='b')
plt.plot(X_test,y_prediksi,color='r')
plt.title("Prediksi Harga, Berdasarkan Lot Area")
plt.xlabel("Luas Tanah")
plt.ylabel("Harga")
```

# MENERAPKAN PREDIKSI 
- Mari prediksi, Pada saat luasnya 12.000,  Berapa Harga Tanah?

```python
plt.figure(figsize=(12,8))
plt.xlabel('Luas Tanah')
plt.ylabel('Price / Harga')
plt.scatter(X_train,y_train)
plt.plot(X_test,y_prediksi, color='r')
plt.axvline(x=12000, color='g')
```
